```python
# creating dialog boxes with tkinter
import tkinter as tk
```

```python
# simple dialog box
def dialog():
    win = tk.Toplevel()
    win.title('Sub menu')
    win.geometry('400x200')
    win.resizable(0, 0)
    tk.Label(win, text='IO box').pack()
    tk.Button(win, text='Return', command=win.destroy).pack()
    win.focus_set()
    win.grab_set()
    win.wait_window()

# test the dialog box
root = tk.Tk()
root.geometry('400x400')
root.title('MAIN UI')
root.resizable(0, 0)
tk.Button(root, text='Enter IO', command=dialog).pack()
root.mainloop()

# creating a custom dialog box

```

```python
from scipy.fft import fft, fftfreq, fftshift
import numpy as np
# number of signal points
N = 400
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N, endpoint=False)
y = np.exp(50.0 * 1.j * 2.0*np.pi*x) + 0.5*np.exp(80.0 * 1.j * 2.0*np.pi*x) + 2
yf = fft(y)
xf = fftfreq(N, T)
yplot = xf

xf = fftshift(xf)

yplot = fftshift(yf)


import matplotlib.pyplot as plt
plt.plot(x,y)
plt.title("input wave")
plt.xlabel("x")
plt.ylabel("amplitude")
plt.show()

plt.plot(xf, (1 / N) * abs(yplot))
plt.grid()
plt.show()
```
